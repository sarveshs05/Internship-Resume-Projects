```python
#some are BHK some are bedroom, standardizing them
df['bhk']=df['size'].str.split().str.get(0).astype(int)
```

```python
df['total_sqft'].unique()
```

```python
#some points have range, so replacing them by their mean
def convertRange(x):
  temp = x.split('-')
  if len(temp) == 2:
    return (float(temp[0])+float(temp[1]))/2
  try:
    return float(x)
  except:
    return None
```

```python
df['total_sqft']=df['total_sqft'].apply(convertRange)
```

```python
df.head()
```

```python
#price/sqft will help in removing outliers
df['price_per_sqft']=df['price']*100000/df['total_sqft']
```

```python
df['location'] = df['location'].apply(lambda x: x.strip())
location_stats=df['location'].value_counts()
location_stats_less_than_10=location_stats[location_stats<=10]
```

```python
df['location'] = df['location'].apply(lambda x: 'other' if x in location_stats_less_than_10 else x)
```

#Outlier Removal


```python
df.describe()
```

```python
(df['total_sqft']/df['bhk']).describe()
```

```python
df=df[((df['total_sqft']/df['bath'])>=300)]
df.describe()
```

```python
df.shape
```

```python
df.price_per_sqft.describe()
```

```python
def remove_outliers_from_sqft(df):
    df_out=pd.DataFrame()
    for key,sub in df.groupby('location'):
        m=np.mean(sub.price_per_sqft)
        std=np.std(sub.price_per_sqft)
        reduce_df=sub[(sub.price_per_sqft>(m-std)) & (sub.price_per_sqft<=(m+std))]
        df_out=pd.concat([df_out,reduce_df],ignore_index=True)
    return df_out
df=remove_outliers_from_sqft(df)
```

```python
df.describe()
```

```python
def bhk_outliers_remover(df):
    exclude_indices=np.array([])
    for location,location_df in df.groupby('location'):
        bhk_stats={}
        for bhk,bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk]={
                'mean':np.mean(bhk_df.price_per_sqft),
                'std':np.std(bhk_df.price_per_sqft),
                'count':bhk_df.shape[0]
            }
        for bhk,bhk_df in location_df.groupby('bhk'):
            stats=bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices=np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')

df=bhk_outliers_remover(df)
```

```python
df = bhk_outliers_remover(df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score, mean_absolute_percentage_error
from sklearn.metrics import mean_squared_error
```

```python
encoder = OneHotEncoder( sparse=False)
df['location']=encoder.fit_transform(df['location'])
```

```python
X=df.drop(columns=['price'])
Y=df['price']
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
print(X_train,X_test.shape)
```

```python
encoder = make_column_transformer((OneHotEncoder(sparse=False),['location']),remainder='passthrough')
scaler = StandardScaler()
lr = LinearRegression()
pipe = make_pipeline(encoder,scaler,lr)
```

```python
pipe.fit(X_train,Y_train)
```

```python
Y_pred_lr=pipe.predict(X_test)
r2_score(Y_test,Y_pred_lr), mean_absolute_percentage_error(Y_test,Y_pred_lr), mean_squared_error(Y_test,Y_pred_lr)
```

```python
lasso = Lasso()
pipe = make_pipeline(encoder,scaler,lasso)
```

```python
pipe.fit(X_train,Y_train)
```

```python
Y_pred_lasso=pipe.predict(X_test)
r2_score(Y_test,Y_pred_lasso), mean_absolute_percentage_error(Y_test,Y_pred_lasso), mean_squared_error(Y_test,Y_pred_lasso)
```

```python
ridge=Ridge()
pipe = make_pipeline(encoder,scaler,ridge)
pipe.fit(X_train,Y_train)
```

```python
Y_pred_ridge=pipe.predict(X_test)
r2_score(Y_test,Y_pred_ridge), mean_absolute_percentage_error(Y_test,Y_pred_ridge), mean_squared_error(Y_test,Y_pred_ridge)
```

```python
print("No Regularization ", r2_score(Y_test,Y_pred_lr), mean_absolute_percentage_error(Y_test,Y_pred_lr), mean_squared_error(Y_test,Y_pred_lr))
print("Lasso ", r2_score(Y_test,Y_pred_lasso), mean_absolute_percentage_error(Y_test,Y_pred_lasso), mean_squared_error(Y_test,Y_pred_lasso))
print("Ridge ", r2_score(Y_test,Y_pred_ridge), mean_absolute_percentage_error(Y_test,Y_pred_ridge), mean_squared_error(Y_test,Y_pred_ridge))
```
